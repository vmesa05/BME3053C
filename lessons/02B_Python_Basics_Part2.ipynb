## [```if``` Statements](https://docs.python.org/3/tutorial/controlflow.html#if-statements)

* Allow you to execute a block of code only if a specified condition is true.
* An ```if``` statement starts with the keyword ```if```, followed by a condition, and a colon ```:```. The indented block of code below it runs if the condition evaluates to True.
```
if condition1:
    print('condition1 is',True)
```
* You can extend if statements with ```else``` for an alternative action if the condition is false, or use ```elif``` (short for "else if") to check additional conditions in sequence.
  * If multiple ```elif``` conditions are True, only the code in the first True ```elif``` will be run
   
```
x = int(input("Please enter an integer: "))
if x < 0:
    print('Input is less than zero')
elif x == 0:
    print('Input equals zero')
elif x > 0:
    print('Input is greater than zero')
else:
    print(ValueError("x is not an integer"))
```

* **Combining Conditions**: The ```and``` operator is used to combine multiple conditions in an if statement, requiring all conditions to be true for the overall condition to be true. For example, ```if condition1 and condition2:``` only executes the block if both condition1 and condition2 are true.
  * ```&``` can be used instead of ```and```.
* **Evaluating Alternatives**: The ```or``` operator allows you to execute a block of code if at least one of the conditions is true. For example, ```if condition1 or condition2:``` will run the block if either condition1 or condition2 is true.
  * ```|``` can be used instead of ```or```.
* **Negating Conditions**: The ```not``` keyword in an ```if``` statement is used to invert a condition, making the block of code execute only if the condition is false. For example, ```if not condition:``` will run the block if condition evaluates to False.

## [```match``` Statements](https://docs.python.org/3/tutorial/controlflow.html#match-statements)

* The ```match``` statements can be used to handle multiple conditions by matching the expression against various patterns defined in ```case``` clauses.

```
def http_error(status):
    match status:
        case 400:
            return "Bad request"
        case 404:
            return "Not found"
        case 418:
```

### Using `in`, `any`, and `all` in Python `if` Statements
* **`in` Keyword**: Checks for membership within a sequence. It's a clean way to see if an item exists in a list, tuple, or string.
  ```python
  symptoms = ['fever', 'cough', 'fatigue']
  if 'cough' in symptoms:
      print("Patient has a cough.")
  ```

* **`any()` Function**: Returns `True` if at least one element in an iterable is true. It's useful for checking if any of several conditions are met.
  ```python
  vital_signs = [98.6, 120, 80] # Temp, Systolic BP, Diastolic BP
  if any(v > 110 for v in vital_signs):
      print("At least one vital sign is elevated.")
  ```

* **`all()` Function**: Returns `True` only if all elements in an iterable are true. This is perfect for ensuring a set of conditions are all satisfied.
  ```python
  pre_op_checklist = [True, True, True] # Consent, Fasting, Vitals Checked
  if all(pre_op_checklist):
      print("Patient is cleared for surgery.")
  ```

````python
def diagnose_blood_pressure(systolic, diastolic):
  """
  Classify blood pressure using match statements.
  
  Args:
    systolic (int): Systolic blood pressure (mmHg)
    diastolic (int): Diastolic blood pressure (mmHg)
  
  Returns:
    str: Blood pressure classification
  """
  
  # Classify based on systolic pressure using match statement
  match systolic:
    case x if x < 90:
      systolic_category = "Low"
    case x if 90 <= x < 120:
      systolic_category = "Normal"
    case x if 120 <= x < 130:
      systolic_category = "Elevated"
    case x if 130 <= x < 140:
      systolic_category = "Stage 1 Hypertension"
    case x if x >= 140:
      systolic_category = "Stage 2 Hypertension"
    case _:
      systolic_category = "Invalid"
  
  # Classify based on diastolic pressure
  match diastolic:
    case x if x < 60:
      diastolic_category = "Low"
    case x if 60 <= x < 80:
      diastolic_category = "Normal"
    case x if 80 <= x < 90:
      diastolic_category = "Stage 1 Hypertension"
    case x if x >= 90:
      diastolic_category = "Stage 2 Hypertension"
    case _:
      diastolic_category = "Invalid"
  
  print(f"Blood Pressure: {systolic}/{diastolic} mmHg")
  print(f"Systolic: {systolic_category}")
  print(f"Diastolic: {diastolic_category}")
  
  # Overall classification (take the higher risk category)
  if "Stage 2" in [systolic_category, diastolic_category]:
    return "Stage 2 Hypertension - Immediate medical attention recommended"
  elif "Stage 1" in [systolic_category, diastolic_category]:
    return "Stage 1 Hypertension - Lifestyle changes and monitoring needed"
  elif systolic_category == "Elevated":
    return "Elevated Blood Pressure - Lifestyle modifications recommended"
  elif "Low" in [systolic_category, diastolic_category]:
    return "Hypotension - Monitor for symptoms"
  else:
    return "Normal Blood Pressure - Continue healthy lifestyle"

# Test with example
print("=== Blood Pressure Classification Demo ===\n")
result = diagnose_blood_pressure(125, 75)
print(f"üìã Diagnosis: {result}")
````

# [Handling Exceptions](https://docs.python.org/3/tutorial/errors.html#handling-exceptions)
 * *try-except* blocks are used to handle exceptions (errors) that may occur during the execution of a program. This helps prevent the program from crashing and allows you to manage errors gracefully.

* The code that might cause an exception is placed inside the try block. If an exception occurs, the program immediately jumps to the corresponding except block, where you can define how to handle the error.

* **Multiple Exception Types**: You can handle different types of exceptions separately by using multiple except blocks, allowing for specific error handling strategies.

```python
try:
  # Code that might raise different types of exceptions
  result = int(user_input) / divisor
except ValueError:
  print("Invalid input: Please enter a number")
except ZeroDivisionError:
  print("Error: Cannot divide by zero")
except Exception as e:
  print(f"Unexpected error: {e}")
```

* **Raising Custom Exceptions**: You can use `raise` to manually trigger exceptions when certain conditions are met, which is useful for validating medical data ranges.

```python
def validate_temperature(temp):
  if temp < 90 or temp > 110:
    raise ValueError(f"Temperature {temp}¬∞F is outside safe range (90-110¬∞F)")
  return temp
```

* **Finally Block**: The `finally` block executes regardless of whether an exception occurred, making it useful for cleanup operations like closing files or releasing resources.

```python
try:
  # Process medical data
  process_patient_data()
except Exception as e:
  print(f"Error processing data: {e}")
finally:
  # Always executed - cleanup operations
  close_database_connection()
```


````python
# Example: Medical Exception Handling
def check_heart_rate(heart_rate):
  """
  Validate heart rate input with medical-specific error handling.
  
  Args:
    heart_rate: Heart rate value to validate
    
  Returns:
    str: Assessment of heart rate
  """
  try:
    # Convert to float in case string input
    hr = float(heart_rate)
    
    # Check for valid medical range
    if hr <= 0:
      raise ValueError("Heart rate must be positive")
    elif hr > 300:
      raise ValueError("Heart rate exceeds physiological maximum")
    
    # Medical assessment
    if hr < 60:
      return f"Bradycardia detected: {hr} bpm"
    elif hr > 100:
      return f"Tachycardia detected: {hr} bpm" 
    else:
      return f"Normal heart rate: {hr} bpm"
      
  except ValueError as e:
    return f"Invalid heart rate input: {e}"
  except Exception as e:
    return f"Unexpected error: {e}"
  finally:
    print("I'm going to run no matter what!")

# Test the function
# Get user input and test the function
user_input = input("Enter a heart rate value: ")
result = check_heart_rate(user_input)
print(f"Result: {result}")
````



 ### The [```range```](https://docs.python.org/3/tutorial/controlflow.html#the-range-function) Function

 Immutable sequence of ordered numbers that is commonly used to loop through code a specific number of times in [```for```](https://docs.python.org/3/reference/compound_stmts.html#for) loops

* The arguments provided when creating a ```range``` object can be used to set a ```start```, a ```stop```, and a ```step``` interval when creating an ordered list of integers
  * Will iterate from
  * ```start```: where the sequence will begin (inclusive)
    * Defaults to 0
  * ```stop```: where the sequence will begin (exclusive)
    * No default value
  * ```step```:
    * defaults to 1
    * cannot be 0
* The number of arguments provided affects the way they are interpreted.
  * ```range(stop)```: Providing a single argument will set the ```stop``` and set ```start``` and ```step``` to their default values.
  *  ```range(start,stop)```: Providing two variables will set ```start``` and ```stop```. ```step``` will be set to 1
    * prividing a ```stop``` that is higher than the  histep and a single argument will set the ```stop``` and set ```start``` and ```step``` to their default values.

````python
for i in range(5):
  print(i)
````

````python
#From: https://docs.python.org/3/library/stdtypes.html#ranges
print(list(range(5)),'= range(5)')

print(list(range(1,5)),'= range(1,5)')
print(list(range(0, 5, 2)),'= range(0, 5, 2)')
print(list(range(0, 10, 3)),'= range(0, 10, 3)')
print(list(range(0, -10, -1)),'= range(0, -10, -1)')
print(list(range(0)), '= range(0)')
print(list(range(4, 2)),'= range(4, 2)')
````

## Continue and Break Statements


* **continue Statement:** Skips the current iteration of a loop and moves directly to the next iteration, effectively ignoring the remaining code within the loop for that iteration.

* **break Statement:** Exits the loop entirely, terminating any further iterations, and proceeds with the code following the loop. It's often used when a certain condition is met and no further processing is needed within the loop.

````python
for i in range(8):
  if i < 2:
    continue
  if i < 4:
    print('i =', i)
    continue
  break
print('final i =',i)
````

### Class and Instance Variables

* Class variables are shared by all instances of a class.
* Instance variables are unique to each instance of a class.


```
class Dog:

    kind = 'canine'         # class variable shared by all instances

    def __init__(self, name):
        self.name = name    # instance variable unique to each instance

>>> d = Dog('Fido')
>>> e = Dog('Buddy')
>>> d.kind                  # shared by all dogs
'canine'
>>> e.kind                  # shared by all dogs
'canine'
>>> d.name                  # unique to d
'Fido'
>>> e.name                  # unique to e
'Buddy'
```

### Mistaken Use of Class Variables

* Class variables are shared by all instances of a class.
* If you mistakenly use a class variable instead of an instance variable, it will be shared by all instances of the class.



````python
class Dog:

    tricks = []             # mistaken use of a class variable

    def __init__(self, name):
        self.name = name

    def add_trick(self, trick):
        self.tricks.append(trick)

d = Dog('Fido')
e = Dog('Buddy')
d.add_trick('roll over')
e.add_trick('play dead')
print(d.tricks)                # unexpectedly shared by
````

````python
class Dog:

    def __init__(self, name):
        self.name = name
        self.tricks = []    # creates a new empty list for each dog

    def add_trick(self, trick):
        self.tricks.append(trick)

d = Dog('Fido')
e = Dog('Buddy')
d.add_trick('roll over')
e.add_trick('play dead')

print(d.tricks)
print(e.tricks)
````
